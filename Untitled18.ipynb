```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:/Users/jeeva/Downloads/a412dfbd88b3db70b74b-5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1 (1)/a412dfbd88b3db70b74b-5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv"  # Update with the correct file path

data = pd.read_csv(file_path)

if 'am' in data.columns:
    transmission_counts = data['am'].value_counts().sort_index()
    transmission_labels = ['Automatic', 'Manual']  # 0 = automatic, 1 = manual

    plt.figure(figsize=(10, 6))
    plt.bar(transmission_labels, transmission_counts, color=['blue', 'green'])
    plt.title('Frequency Distribution of Transmission Types')
    plt.xlabel('Transmission Type')
    plt.ylabel('Frequency')

    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
else:
    print("The 'am' column is not in the dataset. Please check the column names.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:/Users/jeeva/Downloads/a412dfbd88b3db70b74b-5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1 (1)/a412dfbd88b3db70b74b-5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv"

data = pd.read_csv(file_path)

if 'mpg' in data.columns:
    plt.figure(figsize=(8, 6))
    plt.boxplot(data['mpg'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'))
    plt.title('Box Plot of mpg')
    plt.xlabel('Miles Per Gallon (mpg)')
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.show()

    five_number_summary = data['mpg'].describe(percentiles=[.25, .5, .75])
    print("Five-number summary:")
    print(five_number_summary[['min', '25%', '50%', '75%', 'max']])
else:
    print("The 'mpg' column is not in the dataset. Please check the column names.")
```
